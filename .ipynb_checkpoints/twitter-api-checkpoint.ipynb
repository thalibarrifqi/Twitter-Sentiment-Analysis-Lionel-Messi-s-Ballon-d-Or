# **Crawling the Data**

We will use tweepy to retrieve data from Twitter in this notebook; the data will be tweets from Twitter users containing specific keywords.

**Libraries**

```python
import pandas as pd
import tweepy
import time
from datetime import datetime
```

**Token**

```python
bearer_token = 'YOUR BEARER TOKEN ON SPOTIFY APP'
```

**Conect to API**

```python
client = tweepy.Client(bearer_token=bearer_token, wait_on_rate_limit=True)
```

**Get the tweets**

```python
# Replace with your own search query
query = 'Messi -is:retweet -https'

# Start time
start = datetime.now()
print('Start at',datetime.now().strftime("%H:%M:%S"))

# Fill the data
tweets_data = []
for i, tweet in enumerate(tweepy.Paginator(
    client.search_recent_tweets, 
    query=query, 
    tweet_fields=['created_at', 'public_metrics', 'lang'],
    user_fields=['name', 'username'],
    expansions='author_id',
    max_results=100).flatten(limit=100)):
    print(tweet.id)
    print(tweet.created_at)
    print(tweet.text)
    print(tweet.data['lang'])
    print(tweet.data['public_metrics']['like_count'])
    print(tweet.data['public_metrics']['retweet_count'])
    print(client.get_user(id=tweet.author_id).data.name)
    print(client.get_user(id=tweet.author_id).data.username)
    
end = datetime.now()
print('Stop at',datetime.now().strftime("%H:%M:%S"))
print(end-start)
```
